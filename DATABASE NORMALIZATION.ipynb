**IMPORT THE MODULE**

```python
import psycopg2
```

**SET autocommit to true**
- This means that after each call, commit that one action and do not open the transaction for any other action.

**MOVING TO 2ND NORMAL FORM**

- We need to break this into 2 tables, album library and song library, The aim is to reduce the number of duplicates we have in album_id.

**MOVING TO 3RD NORMAL FORM**
###### Our tables still have transitive functional dependency, They are not in 3NF
###### naija_top5_albums can now move artist_name to its own table called Artists, which will leave us with 3 tables.
###### CREATE THE 3 TABLES

- **TABLE NAME** : naija_top5_album_library
- **column 0**   : album_id
- **column 1**   : album_name
- **column 2**   : artist_id
- **column 3**   : release_date






- **TABLE NAME** : naija_top5_song_library
- **column 0**   : song_id
- **column 1**   : song_name
- **column 2**   : album_id






- **TABLE NAME** : naija_top5_artist_library
- **column 0**   : artist_id
- **column 1**   : artist_name


```python
try:
    cur.execute('CREATE TABLE IF NOT EXISTS naija_top5_artist_library(artist_id int PRIMARY KEY, \
                                                                      artist_name varchar(200))')
except Exception as error:
    print(error)

try:
    cur.execute('CREATE TABLE IF NOT EXISTS naija_top5_album_library(album_id int PRIMARY KEY, \
                                                                      album_name varchar(200), \
                                                                      artist_id int, \
                                                                      release_date date, \
                                                                      FOREIGN KEY(artist_id) \
                                                                      REFERENCES naija_top5_artist_library(artist_id))')
except Exception as error:
    print(error)

try:
    cur.execute('CREATE TABLE IF NOT EXISTS naija_top5_song_library(song_id int PRIMARY KEY, \
                                                                    song_name varchar(200), \
                                                                    album_id int, \
                                                                    FOREIGN KEY(album_id) \
                                                                    REFERENCES naija_top5_album_library(album_id))')
except Exception as error:
    print(error)
```
